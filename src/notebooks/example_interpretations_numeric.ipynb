# GPT-4 interpretations

```python
import os
import json
from IPython.display import Image as Ipymage
```

## Numeric

```python
gt_dir = '../find_dataset/numeric/'
est_dir = '../run_interpretations/results/gpt-4/numeric/'
num_func = 100

for func_inx in range(num_func):
    func = f'f{func_inx:05d}'    
    gt_func_dir = os.path.join(gt_dir,func)
    est_func_dir = os.path.join(est_dir,func)
    if not(os.path.exists(os.path.join(est_func_dir,'description.txt'))): continue
    if not(os.path.isdir(gt_func_dir)): continue
    
    
    f = open(os.path.join(gt_func_dir,'function_code.py'))
    gt_func = f.read()    
    gt_func = gt_func.rsplit("import sys")[1]
    gt_func = gt_func.rsplit("if __name__")[0]

    f = open(os.path.join(est_func_dir,'description.txt'))
    est_desc = f.read()
        
    f = open(os.path.join(est_func_dir,'code.py'))
    est_func = f.read()
    
    if est_func == '': continue
        
    print(func,'\n')
    print('*gt code:*\n', gt_func, '\n')
    print('*estimated description:*\n', est_desc, '\n')
    print('*estimated code:*\n', est_func, '\n')    
    try:
        display(Ipymage(filename=est_func_dir + '/test/test.png'))
    except:
        continue

    
```
