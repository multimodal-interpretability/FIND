## Numeric

```python
import os
import json
from IPython.display import Image as Ipymage
```

```python
gt_dir = '../find_dataset/numeric/'
est_dir = '../run_interpretations/results/gpt-4/numeric/'

for func in os.listdir(est_dir):
    gt_func_dir = os.path.join(gt_dir,func)
    est_func_dir = os.path.join(est_dir,func)
    if not(os.path.exists(os.path.join(est_func_dir,'description.txt'))): continue
    if not(os.path.isdir(gt_func_dir)): continue
    
    
    f = open(os.path.join(gt_func_dir,'function_code.py'))
    gt_func = f.read()    
    gt_func = gt_func.rsplit("import sys")[1]
    gt_func = gt_func.rsplit("if __name__")[0]

    f = open(os.path.join(est_func_dir,'description.txt'))
    est_desc = f.read()
        
    f = open(os.path.join(est_func_dir,'code.py'))
    est_func = f.read()
    
    if est_func == '': continue
        
    print(func,'\n')
    print('*gt code:*\n', gt_func, '\n')
    print('*estimated description:*\n', est_desc, '\n')
    print('*estimated code:*\n', est_func, '\n')    
    try:
        display(Ipymage(filename=est_func_dir + '/test/test.png'))
    except:
        continue

    
```

## Synthetic neurons entities

```python
gt_dir = '../find_dataset/neurons_entities/'
est_dir = '../run_interpretations/results/gpt-4/neurons_entities/'

for func in os.listdir(est_dir):
    gt_func_dir = os.path.join(gt_dir,func)
    est_func_dir = os.path.join(est_dir,func)
    if not(os.path.exists(os.path.join(est_func_dir,'description.txt'))): continue
    if not(os.path.isdir(gt_func_dir)): continue
    
    
    f = open(os.path.join(gt_func_dir,'function_code.py'))
    gt_func = f.read()    
    
    if int(func[1:])<140:
        gt_con = gt_func.rsplit("concept = '")[1]
        gt_con = gt_con.rsplit("'\n\ndef function")[0]    
        gt_prompt = 'I am going to give you an input. Return only 1 if that input is related to {concept}. Return only 0 otherwise.\nDo not return any additional information other than the 0 or 1 answer.'
        gt_prompt = gt_prompt.replace('{concept}',gt_con)
    else:
        gt_prompt = gt_func.rsplit(f"I am going to")[1]
        gt_prompt = 'I am going to' + gt_prompt.rsplit('f"Input: {input}"}')[0]
    f = open(os.path.join(est_func_dir,'description.txt'))
    est_desc = f.read()
    
    print(func,'\n')
    print('*gt code:*\n', gt_prompt , '\n')
    print('*estimated description:*\n', est_desc, '\n')


    
```

## Synthetic neurons relations

```python
gt_dir = '../find_dataset/neurons_relations/'
est_dir = '../run_interpretations/results/gpt-4/neurons_relations/'

for func in os.listdir(est_dir):
    gt_func_dir = os.path.join(gt_dir,func)
    est_func_dir = os.path.join(est_dir,func)
    if not(os.path.exists(os.path.join(est_func_dir,'description.txt'))): continue
    if not(os.path.isdir(gt_func_dir)): continue
    
    
    f = open(os.path.join(gt_func_dir,'function_code.py'))
    gt_func = f.read()    
    
    if int(func[1:])<25:
        gt_con1 = gt_func.rsplit("con1 = '")[1]
        gt_con1 = gt_con1.rsplit("'\ncon2 = ")[0] 
        gt_con2 = gt_func.rsplit("con2 = '")[1]
        gt_con2 = gt_con2.rsplit("'\n\ndef function")[0]    
        gt_prompt = "I am going to give you an input. If the input is {con1}, reply with only {con2}. If the input is not {con1}, return 'undefined'.\n Reply with a single word. Do not include any additional information other than the output word.\n"
        gt_prompt = gt_prompt.replace('{con1}',gt_con1).replace('{con2}',gt_con2)
    else:
        gt_con1 = gt_func.rsplit("con1 = '")[1]
        gt_con1 = gt_con1.rsplit("'\ncon2 = ")[0] 
        gt_con2 = gt_func.rsplit("con2 = '")[1]
        gt_con2 = gt_con2.rsplit("'\nex = '")[0] 
        ex = gt_func.rsplit("ex = '")[1]
        ex = ex.rsplit("'\n\ndef function")[0]    
        gt_prompt = "I am going to give you an input. If the input is {con1}, reply with only {con2}. If the input is not {con1}, return 'undefined'. If {ex}, return 'undefined'.\nReply with a single word. Do not include any additional information other than the output word.\n"
        gt_prompt = gt_prompt.replace('{con1}',gt_con1).replace('{con2}',gt_con2).replace('{ex}',ex)
    
    f = open(os.path.join(est_func_dir,'description.txt'))
    est_desc = f.read()
    
    print(func,'\n')
    print('*gt code:*\n', gt_prompt , '\n')
    print('*estimated description:*\n', est_desc, '\n')


    
```
